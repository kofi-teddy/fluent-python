```python
import numpy

# find the mean of the array
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.mean(speed)
print(x)
```

```python
# find the median of the array

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = numpy.median(speed)
x
```

```python
from scipy import stats

# find the frequency of the array
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = stats.mode(speed)
x
```

```python
import numpy

# Standard deviation is a number that describes how spread out the values are.
# A low standard deviation means that most of the numbers are close to the mean (average) value.
# A high standard deviation means that the values are spread out over a wider range.

speed = [86,87,88,86,87,85,86]

x = numpy.std(speed)
x

```

```python
# Variance 
# is another number that indicates how spread out the values are.
# if you take the square root of the variance, you get the standard deviation.
# if you multiply the standard deviation by itself, you get the variance.
# To calculate the variance you have to do as follows:
# 1. Find the mean
# 2. For each value: find the difference from the mean
# 3. For each difference: find the square value
# 4. The variance is the average number of these squared differences

speed = [32,111,138,28,59,77,97]
x = numpy.var(speed)
x
```

```python
# Percentiles 
# describes the value that a given percent of the values are lower than.
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
x = numpy.percentile(ages, 75)
print(x)
x = numpy.percentile(ages, 90)
print(x)
```

```python
# Regression
# used when you try to find the relationship between variables.
# used to predict the outcome of future events

# Linear Regression
# uses the relationship between the data-points to draw 
# a straight line through all them
# This line can be used to predict future values

from scipy import stats


x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
# R for Relationship
# This relationship - the coefficient of correlation - is called r.
# The r value ranges from -1 to 1, where 0 means no relationship,
# and 1 (and -1) means 100% related.

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

print(r)
```

```python
# Polynomial Regression
# If data points clearly does not fit a 
# linear regression (a straight line through all data points), 
# it might be ideal for polynomial regression.
# uses the relationship between the variables x and y 
# to find the best way to draw a line through the data points.

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))
myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
# Multiple regression 
# a linear regression, but with more than one independent value, 
# meaning that we try to predict a value based on two or more 
# variables.

import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np


# Load the CSV file
df = pd.read_csv("data/car-emission.csv")

# Select features and target
X = df[['Weight', 'Volume']]
y = df['CO2']

# Initialize and fit the model
model = LinearRegression()
model.fit(X, y)

# # predict the CO2 emission of a car where the weight 
# # is 2300kg, and the volume is 1300cm3:
predictions = model.predict([[2300, 1300]])
# new_data = np.array([[1500, 1200], [1200, 1000]])
# predictions = model.predict(new_data)

print(predictions)


```

```python
# Scale Features
# When your data has different values, and even different 
# measurement units, it can be difficult to compare them. 
# What is kilograms compared to meters? Or altitude 
# compared to time?

# Standardization method
# The standardization method uses this formula:
# z = (x - u) / s
# Where z is the new value, x is the original value, 
# u is the mean and s is the standard deviation.

from sklearn.preprocessing import StandardScaler


scale = StandardScaler()
X = df[['Weight', 'Volume']]
scaledX = scale.fit_transform(X)
print(scaledX)
```

```python
# Predict the CO2 emission from a 1.3 
# liter car that weighs 2300 kilograms:


X = df[['Weight', 'Volume']]
y = df['CO2']

scaledX = scale.fit_transform(X)

model = LinearRegression()
model.fit(scaledX, y)

scaled = scale.transform([[2300, 1.3]])

predictions = model.predict([scaled[0]])
print(predictions)

```

```python
# Train/Test
# measure the accuracy of your model
# It is called Train/Test because you split the data set 
# into two sets: a training set and a testing set

np.random.seed(2)

x = np.random.normal(3, 1, 100)
y = np.random.normal(150, 40, 100) / x

plt.scatter(x, y)
plt.show()
```

```python
# The training set should be a random selection 
# of 80% of the original data.

# The testing set should be the remaining 20%.

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]
```

```python
# Draw a polynomial regression line through the data points:

model = np.poly1d(numpy.polyfit(train_x, train_y, 4))
line = np.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(line, model(line))
plt.show()
```

```python
# R-squared, R2
# It measures the relationship between the x axis 
# and the y axis, and the value ranges from 0 to 1, 
# where 0 means no relationship, and 1 means totally related.

r2 = r2_score(train_y, model(train_x))
print(r2)
```

```python
# R2 score when using testing data

np.random.seed(2)

x = np.random.normal(3, 1, 100)
y = numpy.random.normal(150, 40, 100) / x

train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]

model = np.poly1d(np.polyfit(train_x, train_y, 4))

r2 = r2_score(test_y, model(test_x))

print(r2)
```
